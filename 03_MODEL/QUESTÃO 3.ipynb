```python
from ortools.linear_solver import pywraplp
```

```python
# declara o solver: se PLI, usa-se SCIP; se PL, usa-se GLOP.
solver = pywraplp.Solver.CreateSolver('SCIP')

# declara um número muito grande caso seja necessário usar no modelo
infinity = solver.infinity()
```

```python
# declara as variáveis do modelo: IntVar para inteira; NumVar para fracionárias; BoolVar para binárias.
x1 = solver.IntVar(3, infinity, 'x1')     # quantidade de quadros
x2 = solver.IntVar(0, infinity, 'x2')     # quantidade de paineis
x3 = x1 * 2;

# solver.IntVar(rangeMinimo, rangeMaximo, 'nome')

# x3 = solver.NumVar(0,0, infinity, 'x3')
# x4 = solver.BoolVar('x3')
```

```python
# declara a função objetivo
# 150.x1 + 250.x2
objetivo = solver.Objective()
objetivo.SetCoefficient(x1, 150) # 150.x1
objetivo.SetCoefficient(x2, 250) # 250.x2
objetivo.SetMaximization()                # SetMinimization para minimização
```

```python
# declara uma restrição
# 8.x1 + 12.x2 <= 180
restricao_1 = solver.RowConstraint(-infinity, 180, 'linha_1') # infinito <= 100      # colocar um nome para a restrição é opcional
restricao_1.SetCoefficient(x1, 8)  # 10*x1
restricao_1.SetCoefficient(x2, 12)  # 10*x2
```

```python
# declara uma restrição
# se executar vezes, gera 2 restrições, tem que reiniciar a execução em ambiente de execução:
# 90.x1 + 140.x2 <= 2000
restricao_2 = solver.RowConstraint(-infinity, 2000, 'linha_2')        # colocar um nome para a restrição é opcional
restricao_2.SetCoefficient(x1, 80)  # 3*x1
restricao_2.SetCoefficient(x2, 140)  # 7*x2
```

```python
# declara uma restrição
# se executar vezes, gera 2 restrições, tem que reiniciar a execução em ambiente de execução:
# x2 >= 2.x1
restricao_3 = solver.RowConstraint(-infinity, 0, 'linha_3')        # colocar um nome para a restrição é opcional
restricao_3.SetCoefficient(x1, 2)  # 3*x1
restricao_3.SetCoefficient(x2, -1)  # 3*x1
```

```python
# resolve o modelo
status = solver.Solve()
```

```python
# verifica se a solução é ótima e, caso seja, exibe ela
if status == pywraplp.Solver.OPTIMAL:
  print('Lucro:', objetivo.Value())
  print(x1.name(), x1.solution_value())
  print(x2.name(), x2.solution_value())
else:
  print('Modelo sem solução ótima.')
```
