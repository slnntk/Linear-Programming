*Exerccio 2 Uma marcenaria deseja estabelecer uma programacao diaria de
 producao. Atualmente, a marcenaria faz apenas dois produtos: mesa e ca
deira. Para efeito de simpli cacao, vamos considerar que a marcenaria tem
 limitacoes em somente dois recursos: madeira e mao-de-obra, cujas dispo
nibilidades diarias sao de 14m2 de madeira e 10 horas de mao-de-obra. O
 processo de producao e tal que, para fazer uma mesa, a fabrica gasta 2m2 de
 madeira e 2 horas de mao-de-obra. Para fazer uma cadeira, a fabrica gasta
 3m2 de madeira e 1 hora de mao-de-obra. Alem disso, o fabricante sabe que
 cada mesa tem um custo de producao igual a R$$ 12 e seu preco de venda e
 R$52. Cada cadeira tem um custo de producao igual a R$ 10 e seu preco de
 venda e R$20. Modele o problema para maximizar o lucro.*

```python
from ortools.linear_solver import pywraplp
```

```python
# declara o solver: se PLI, usa-se SCIP; se PL, usa-se GLOP.
solver = pywraplp.Solver.CreateSolver('SCIP')

# declara um número muito grande caso seja necessário usar no modelo
infinity = solver.infinity()
```

```python
# declara as variáveis do modelo: IntVar para inteira; NumVar para fracionárias; BoolVar para binárias.
x1 = solver.IntVar(0, infinity, 'x1')     # quantidade de quadros
x2 = solver.IntVar(0, infinity, 'x2')     # quantidade de paineis

# solver.IntVar(rangeMinimo, rangeMaximo, 'nome')

# x3 = solver.NumVar(0,0, infinity, 'x3')
# x4 = solver.BoolVar('x3')
```

```python
# declara a função objetivo
# 40.x1 + 10.x2
objetivo = solver.Objective()
objetivo.SetCoefficient(x1, 40) # 40.x1
objetivo.SetCoefficient(x2, 10) # 10.x2
objetivo.SetMaximization()                # SetMinimization para minimização
```

```python
# declara uma restrição
# 2.x1 + 3.x2 <= 14
restricao_1 = solver.RowConstraint(-infinity, 14, 'linha_1')       # colocar um nome para a restrição é opcional
restricao_1.SetCoefficient(x1, 2)  # 2*x1
restricao_1.SetCoefficient(x2, 3)  # 3*x2
```

```python
# declara uma restrição
# 2.x1 + 1.x2 <= 10
restricao_2 = solver.RowConstraint(-infinity, 10, 'linha_2')        # colocar um nome para a restrição é opcional
restricao_2.SetCoefficient(x1, 2)  # 2*x1
restricao_2.SetCoefficient(x2, 1)  # 1*x2
```

```python
# resolve o modelo
status = solver.Solve()
```

```python
# verifica se a solução é ótima e, caso seja, exibe ela
if status == pywraplp.Solver.OPTIMAL:
  print('Lucro:', objetivo.Value())
  print(x1.name(), x1.solution_value())
  print(x2.name(), x2.solution_value())
else:
  print('Modelo sem solução ótima.')
```

```python
# exibe o modelo
print(solver.ExportModelAsLpFormat(False))
```
