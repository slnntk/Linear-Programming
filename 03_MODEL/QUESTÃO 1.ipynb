*Exerccio 1 A Esportes Radicais S/A produz para-quedas e asa-deltas em
 duas linhas de montagem. A primeira linha de montagem tem 100 horas se
manais disponveis para a fabricacao dos produtos, e a segunda linha tem um
 limite de 42 horas semanais. Cada um dos produtos requer 10 horas de pro
cessamento na linha 1, enquanto que na linha 2 o para-quedas requer 3 horas
 e a asa-delta requer 7 horas. Sabendo que o mercado esta disposto a comprar
 toda a producao da empresa e que o lucro pela venda de cada para-quedas e
 de R$$60 e para cada asa-delta vendida e de R$40, encontre a programacao
 de producao que maximize o lucro da Esportes Radicais S/A.*

```python
from ortools.linear_solver import pywraplp
```

```python
# declara o solver: se PLI, usa-se SCIP; se PL, usa-se GLOP.
solver = pywraplp.Solver.CreateSolver('SCIP')

# declara um número muito grande caso seja necessário usar no modelo
infinity = solver.infinity()
```

```python
                                                                                  # declara as variáveis do modelo: IntVar para inteira; NumVar para fracionárias; BoolVar para binárias.
x1 = solver.IntVar(0, infinity, 'x1')     # quantidade de paraquedas
x2 = solver.IntVar(0, infinity, 'x2')     # quantidade de asa-deltas

                                                                                  # solver.IntVar(rangeMinimo, rangeMaximo, 'nome')

                                                                                  # x3 = solver.NumVar(0,0, infinity, 'x3')
                                                                                  # x4 = solver.BoolVar('x3')
```

```python
# declara a função objetivo
# 60.x1 + 40.x2
objetivo = solver.Objective()
objetivo.SetCoefficient(x1, 60) # 60.x1
objetivo.SetCoefficient(x2, 40) # 40.x2
objetivo.SetMaximization()                # SetMinimization para minimização
```

```python
# declara uma restrição
# 10.x1 + 10.x2 <= 100
restricao_1 = solver.RowConstraint(-infinity, 100, 'primeira linha de montagem') # infinito <= 100      # colocar um nome para a restrição é opcional
restricao_1.SetCoefficient(x1, 10)  # 10*x1
restricao_1.SetCoefficient(x2, 10)  # 10*x2
```

```python
# declara uma restrição
# 3.x1 + 7.x2 <= 42
restricao_2 = solver.RowConstraint(-infinity, 42, 'segunda linha de montagem')        # colocar um nome para a restrição é opcional
restricao_2.SetCoefficient(x1, 3)  # 3*x1
restricao_2.SetCoefficient(x2, 7)  # 7*x2
```

```python
# resolve o modelo
status = solver.Solve()
```

```python
# verifica se a solução é ótima e, caso seja, exibe ela
if status == pywraplp.Solver.OPTIMAL:
  print('Lucro:', objetivo.Value())
  print(x1.name(), x1.solution_value())
  print(x2.name(), x2.solution_value())
else:
  print('Modelo sem solução ótima.')
```

```python
# exibe o modelo
print(solver.ExportModelAsLpFormat(False))
```
